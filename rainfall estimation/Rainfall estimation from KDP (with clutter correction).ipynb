# Rainfall estimation from Specific Differential Phase (KDP) (with clutter correction)

As an important pre-processing step, the clutter identification and correction is also included in this notebook. This means that all the other polarimetric variables, not just the KDP, is needed. If the clutter correction step is not necessary, or is wished to be skipped, then kindly check the other notebook on Rainfall Estimation from Specific Differential Phase (KDP) (without clutter correction)

```python
%pylab inline
import numpy as np
import pylab as pl
import wradlib as wrl

import warnings
warnings.filterwarnings("ignore")
```

```python
# read the data
data_phi, attrs_phi = wrl.io.read_EDGE_netcdf('../phirad-data/radars/tagaytay/TAG-20120806-131547-02-P.nc')
data_zdr, attrs_zdr = wrl.io.read_EDGE_netcdf('../phirad-data/radars/tagaytay/TAG-20120806-131547-02-D.nc')
data_rho, attrs_rho = wrl.io.read_EDGE_netcdf('../phirad-data/radars/tagaytay/TAG-20120806-131547-02-R.nc')
data_v, attrs_v = wrl.io.read_EDGE_netcdf('../phirad-data/radars/tagaytay/TAG-20120806-131547-02-V.nc')
data_z, attrs_z = wrl.io.read_EDGE_netcdf('../phirad-data/radars/tagaytay/TAG-20120806-131547-02-Z.nc')

# read static clutter map
cmap, attrs_cmap = wrl.io.from_hdf5('../phirad-data/cluttermask/TAG_cmask.hdf5')

# create dictionary
dat = {}
dat["phi"] = data_phi
dat["zdr"] = data_zdr
dat["rho"] = data_rho
dat["dop"] = data_v
dat["map"] = cmap

clutter, nanmask = wrl.clutter.classify_echo_fuzzy(dat, thresh=0.5)

mask = np.isnan(data_z)
clutter_masked = np.ma.masked_array(clutter[0], mask=mask)
```

```python
# mark clutter in phidp using NaN (is filled in the next step)
dat["phi"][clutter[0]] = np.nan
dat["phi"][nanmask] = np.nan
```

```python
dat["phi"], kdp = wrl.dp.process_raw_phidp_vulpiani(dat["phi"], dr=0.5, L=5, niter=4)
```

```python
RKDP = wrl.trafo.kdp2r(kdp, f=1, a=34.6, b=0.83)
```

```python
RKDP[np.any(clutter, axis=0)] = np.nan
RKDP_interp = wrl.ipol.interpolate_polar(np.ma.masked_invalid(RKDP))
```

```python
pl.figure(figsize=(8,8))
wrl.vis.plot_ppi(np.ma.masked_invalid(RKDP_interp),vmin=0, vmax=100)
```
