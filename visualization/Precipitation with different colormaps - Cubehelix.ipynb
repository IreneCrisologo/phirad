# Precipitation plot with different colormaps - Cubehelix

A notebook showcasing precipitation plots using different colormaps from the Cubehelix palettes, described [here](https://jiffyclub.github.io/palettable/cubehelix/)

```python
%pylab inline
import numpy as np
import pylab as pl
import wradlib as wrl
import palettable.cubehelix as pal

import warnings
warnings.filterwarnings("ignore")
```

Read precipitation data.

```python
filename = wrl.util.get_wradlib_data_file('misc/polar_dBZ_tur.gz')
data = np.loadtxt(filename)
```
